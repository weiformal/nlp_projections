```python
from langchain.chains import LLMMathChain
import json
from langchain.utilities import SerpAPIWrapper
import os 
import sentence_transformers
from langchain.vectorstores.chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
%run utils.ipynb
os.environ["SERPAPI_API_KEY"] = "aa75b60f3ae9dd286759062b9c2570fb6255c63c9678067617ad8628a7a4b1f4"
search = SerpAPIWrapper()


from langchain.pydantic_v1 import BaseModel, Field
from langchain.agents.tools import Tool


def query_materical_type(a):
    """查询所有物资类型"""
    sql = "select 物资名称 from 物资表"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

class materical_name(BaseModel):
    mat_type: str = Field(description="物资类型")
    
def query_materical_name(mat_type):
    mat_type = mat_type.split("'")[1] if "'" in mat_type else mat_type
    """根据物资类型查询该类型具体的物品名称"""
    
    sql = f"select distinct {mat_type}名称 from {mat_type}"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

class materical_info(BaseModel):
    mat_dict: str = Field(description="{mat_name:物资名称,mat_type:物资类型}")
    

def query_materical_info(mat_dict):
    """
    根据物资类型和物品名称以及时间在sql表中查询该物品的当前时间的信息状态
    该sql表的字段为[`id`, `名称`, `物资类型`, `库存量`, `消耗量`, `补给量`, `阈值`, `时间`]
    """
    try:
        input = json.loads(find_dict(mat_dict))
    except:
        return "查询失败，输入格式不正确，请按字典格式输出"
    mat_name = input["mat_name"].split("'")[1] if "'" in input['mat_name'] else input['mat_name']
    mat_type = input["mat_type"].split("'")[1] if "'" in input['mat_type'] else input['mat_type']
    sql = f"select * from {mat_type} where {mat_type}名称 = '{mat_name}' limit 20"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

class cabin_info(BaseModel):
    cabin_name: str = Field(description="舱室名称")
    
def query_cabin_info(cabin_name):

    """
    根据舱室名称可以在sql表中查询该舱室的所有信息
    sql表的字段为[`id`, `舱室名称`, `温度`, `湿度`, `温度阈值`, `湿度阈值`, `空调模式`, `风速`, `时间`]
    """
    cabin_name = cabin_name.split("'")[1] if "'" in cabin_name else cabin_name
    sql = f"select * from 舱室 where 舱室名称 = '{cabin_name}' limit 5"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

class position_info(BaseModel):
    position_name: str = Field(description="职位名称")
    
def query_position_info(position_name):

    """
    根据职位名称可以在sql表中查询该职位的职责
    """
    position_name = position_name.split("'")[1] if "'" in position_name else position_name
    sql = f"select 职责 from 船员职责 where 职位 = '{position_name}' limit 5"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

class cabin_except(BaseModel):
    val_dict: str = Field(description='{"val":值,"val_threshold":相应的阈值}')

def der_cabin_except(val_dict):

    """
    该函数可以判断舱室温度或湿度是否异常，判定规则为：
    如果温度大于温度阈值则温度异常，否则为正常
    如果湿度大于湿度阈值则湿度异常，否则为正常
    """
    try:
        input = json.loads(find_dict(val_dict))
    except:
        return '查询失败，输入格式不正确，请按字典格式输出{"val":值,"val_threshold":阈值}'
    val = input['val']
    val_threshold = input['val_threshold']
    if int(val) > int(val_threshold):
        return "该值大于阈值，存在异常"
    else:
        return "该值小于阈值，正常"



def query_device_type(a):
    """查询所有设备的类型"""
    sql = "select 设备名称 from 设备"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

class device_all_info(BaseModel):
    device_name: str = Field(description="device_name:设备名称")
    
def query_device_all_info(device_name):
    """
    根据设备名称，在sql表中查询该设备的当前时间的所有信息状态
    不同设备类型的字段不同
    [id`, `设备名称`, `设备型号`, `部署位置`, `尺寸`, `材料`, `风量`, `风速`, `风压`, `阻力`, `湿度`, `静压损失`, `动压损失`, `噪音级别`, `漏风率`, `绝缘效能`, `维护周期`, `维护要求`, `时间`]
    [`id`, `设备名称`, `部署位置`, `设备型号`, `制冷能力`, `制热能力`, `空气流量`, `温度阈值`, `温度`, `风速控制`, `制冷剂类型`, `风机类型`, `控制方式`, `电压`, `频率`, `噪音级别`, `节能性能`, `故障诊断和报警系统`, `安全特性`, `维护周期`, `维护要求`, `时间`]
    """
    device_name = device_name.split("'")[1] if "'" in device_name else device_name
    sql = f"select * from {device_name[0:4]} where 设备名称 = '{device_name}' limit 5"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res

def search_weather(query : str):
    """这是一个查询天气的工具"""
    
    return search.run(query)
    
def query_department(a):
    """这是一个查询部门与部门人数的工具"""
    sql = "select * from 部门"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res
    
def query_attendance(a):
    """这是一个查询人员出勤率的工具"""
    sql = "select * from 出勤"
    cursor.execute(sql)
    res = cursor.fetchall()
    return res
    
class device_operate_book(BaseModel):
    query: str = Field(description="查询设备类的问题")
    
def search_operate_book(query:str):
    """这是一个可以和查询设备操作手册的工具，如果查询设备类问题，没有其他工具能够查询到答案，可以使用该工具"""
    embedding_model = "/home/weibs/pythonproject/jupyter/study_langchain/localmodel/bce-embedding-base_v1"
    embeddings = HuggingFaceEmbeddings(model_name=embedding_model, model_kwargs={'device': "cuda:0"})
    embeddings.client = sentence_transformers.SentenceTransformer(embeddings.model_name, device="cuda:0")
    documents = Chroma(persist_directory='chroma_db_demo', embedding_function=embeddings)
    # docs = documents.similarity_search('人机交互单元是什么?')
    docs = documents.as_retriever(search_type="mmr", search_kwargs={'k': 3})
    return docs.invoke(query)[0].page_content
```
